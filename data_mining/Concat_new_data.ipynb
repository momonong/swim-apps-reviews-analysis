```python
import pandas as pd
```

```python
def concat_data(app_name: str) -> pd.DataFrame:
    # 讀取CSV文件
    google_reviews = pd.read_csv(f'{app_name}_GooglePlayStore.csv')
    apple_reviews = pd.read_csv(f'{app_name}_apple.csv')

    # 重命名欄位以對齊兩個數據集的欄位名稱
    google_reviews.rename(columns={
        'content': 'review',
        'score': 'rating',
        'thumbsUpCount': 'thumbs_up_count',
        'reviewCreatedVersion': 'review_version',
        'at': 'date',
        'appVersion': 'app_version',
        'platform': 'platform'
    }, inplace=True)

    apple_reviews.rename(columns={
        'date': 'date',
        'review': 'review',
        'rating': 'rating',
        'title': 'title',
        'platform': 'platform'
    }, inplace=True)

    # 選擇各數據集中相關的欄位
    google_reviews = google_reviews[['review', 'rating', 'thumbs_up_count', 'review_version', 'date', 'app_version', 'platform']]
    apple_reviews = apple_reviews[['date', 'review', 'rating', 'title', 'platform']]

    # 為了對齊兩個數據集的欄位，為缺少的欄位添加空值
    apple_reviews['thumbs_up_count'] = None
    apple_reviews['review_version'] = None
    apple_reviews['app_version'] = None
    google_reviews['title'] = None

    # 合併兩個數據集
    combined_reviews = pd.concat([google_reviews, apple_reviews], ignore_index=True, sort=False)
    
    combined_reviews.to_csv(f'{app_name}_concat.csv')

    return combined_reviews
```

```python
def concat_duoswim(app_name='duoswim') -> pd.DataFrame:
    # 讀取CSV文件
    apple_reviews = pd.read_csv(f'{app_name}_apple.csv')

    # 重命名欄位以對齊兩個數據集的欄位名稱
    apple_reviews.rename(columns={
        'date': 'date',
        'review': 'review',
        'rating': 'rating',
        'title': 'title',
        'platform': 'platform'
    }, inplace=True)

    # 選擇各數據集中相關的欄位
    apple_reviews = apple_reviews[['date', 'review', 'rating', 'title', 'platform']]

    # 為了對齊兩個數據集的欄位，為缺少的欄位添加空值
    apple_reviews['thumbs_up_count'] = None
    apple_reviews['review_version'] = None
    apple_reviews['app_version'] = None

    # 合併兩個數據集    
    apple_reviews.to_csv(f'{app_name}_concat.csv')

    return apple_reviews
```

```python
duoswim_concat = concat_duoswim()
form_concat = concat_data('form-swim')
myswimpro_concat = concat_data('myswimpro')
swimcom_concat = concat_data('swim-com')
swimup_concat = concat_data('swimup')
```
