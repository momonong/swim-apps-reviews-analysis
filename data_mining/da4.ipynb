```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_csv('new_result/all_concat_sentiment.csv')
```

```python
data['review'] = data['review'].astype(str)

# 提取 duoswim 的評論
duoswim_reviews = data[data['app'] == 'duoswim']['review']

# 定義功能詞彙
features = {
    'Training Plans & Workouts': ['training plans', 'workouts', 'plans', 'workout plans'],
    'Coaching & Instruction': ['coaching', 'video', 'videos', 'technique'],
    'Data Tracking & Integration': ['track', 'tracking', 'data', 'integration', 'Apple Watch'],
    'Nutrition & Health': ['nutrition', 'diet', 'recipes'],
    'User Experience & Interface': ['easy to use', 'interface', 'layout'],
    'Suggestions for Improvement': ['improvement', 'improve', 'suggestion', 'customize']
}

# 計算每個功能詞彙的出現頻率
feature_counts = {feature: 0 for feature in features}

for review in duoswim_reviews:
    for feature, keywords in features.items():
        if any(keyword in review.lower() for keyword in keywords):
            feature_counts[feature] += 1

# 將結果轉換為 DataFrame
df = pd.DataFrame(list(feature_counts.items()), columns=['Feature', 'Count'])

# 繪製柱狀圖
plt.figure(figsize=(8, 4))
bars = plt.barh(df['Feature'], df['Count'], color='lightcoral')

# 添加具體數值
for bar in bars:
    plt.text(bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, 
             f'{bar.get_width():.0f}', 
             va='center', ha='center', fontsize=10, color='black')

plt.xlabel('Count')
plt.ylabel('Feature')
plt.title('Duoswim User Reviews Analysis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 檔案
df = pd.read_csv('new_result/all_concat.csv')

# 確保 'date' 欄位是 datetime 類型
df['date'] = pd.to_datetime(df['date'])

# 找出每個 app 的評論日期區間
date_ranges = df.groupby('app')['date'].agg(['min', 'max']).reset_index()

# 繪製日期區間的視覺化
plt.figure(figsize=(8, 4))

# 繪製每個 app 的日期區間
for index, row in date_ranges.iterrows():
    plt.plot([row['min'], row['max']], [index, index], marker='o', linestyle='-', color='darkblue', linewidth=3)

# 設置 y 軸的標籤為 app 名稱
plt.yticks(range(len(date_ranges)), date_ranges['app'])

# 添加標題和標籤
plt.title('Review Date Ranges for Different Apps')
plt.xlabel('Date')
plt.ylabel('App')

# 顯示圖表
plt.show()
```

```python
# 生成箱线图
plt.rcParams['figure.figsize'] = [6, 6.5]
boxplot = df.boxplot(column='subjectivity', by='app', grid=False)

plt.title('User Subjectivity for Different Swimming Training Apps')
plt.xlabel('App')
plt.ylabel('Subjectivity')
plt.suptitle('')
plt.xticks(rotation=0)  # 确保应用程序名称是水平的

# 获取每个箱线图的五数总结数据（最小值、第一四分位数、中位数、第三四分位数、最大值）
grouped = df.groupby('app')['subjectivity']
stats = grouped.describe()

adjustments = {
    'duoswim': 0.04,
    'form-swim': 0.06,
    'myswimpro': 0.06,
    'swim-com': 0.06,
    'swimup': 0.06
}

for i, app in enumerate(stats.index, 1):  # enumerate 从1开始，因为箱线图的箱子从1开始
    mean = stats.loc[app, 'mean']
    adjustment = adjustments[app]
    
    # 调整标签的位置和字体
    plt.text(i, mean + adjustment, f'{mean:.2f}', horizontalalignment='center', verticalalignment='top', fontsize=10, color='green', bbox=dict(facecolor='none', edgecolor='none', alpha=0.5))

# 添加显著性标记
def add_significance(ax, x1, x2, y, text):
    x = [x1, x1, x2, x2]
    y_list = [y, y + 0.01, y + 0.01, y]
    ax.plot(x, y_list, lw=1.5, color='black')
    ax.text((x1 + x2) * .5, y + 0.015, text, ha='center', va='bottom', color='black', fontsize=10)

ax = plt.gca()
y_max = df['subjectivity'].max() + 0.1
add_significance(ax, 2, 3, y_max, '*')
add_significance(ax, 2, 4, y_max + 0.05, '**')

plt.show()
```
