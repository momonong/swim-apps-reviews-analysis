```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
data = pd.read_csv('new_result/all_concat_sentiment.csv')
```

```python
data['review'] = data['review'].astype(str)

# 提取 duoswim 的評論
duoswim_reviews = data[data['app'] == 'duoswim']['review']

# 定義功能詞彙
features = {
    'Training Plans & Workouts': ['training plans', 'workouts', 'plans', 'workout plans'],
    'Coaching & Instruction': ['coaching', 'video', 'videos', 'technique'],
    'Data Tracking & Integration': ['track', 'tracking', 'data', 'integration', 'Apple Watch'],
    'Nutrition & Health': ['nutrition', 'diet', 'recipes'],
    'User Experience & Interface': ['easy to use', 'interface', 'layout'],
    'Suggestions for Improvement': ['improvement', 'improve', 'suggestion', 'customize']
}

# 計算每個功能詞彙的出現頻率
feature_counts = {feature: 0 for feature in features}

for review in duoswim_reviews:
    for feature, keywords in features.items():
        if any(keyword in review.lower() for keyword in keywords):
            feature_counts[feature] += 1

# 將結果轉換為 DataFrame
df = pd.DataFrame(list(feature_counts.items()), columns=['Feature', 'Count'])

# 繪製柱狀圖
plt.figure(figsize=(8, 4))
bars = plt.barh(df['Feature'], df['Count'], color='lightcoral')

# 添加具體數值
for bar in bars:
    plt.text(bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, 
             f'{bar.get_width():.0f}', 
             va='center', ha='center', fontsize=10, color='black')

plt.xlabel('Count')
plt.ylabel('Feature')
plt.title('Duoswim User Reviews Analysis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 檔案
df = pd.read_csv('new_result/all_concat.csv')

# 確保 'date' 欄位是 datetime 類型
df['date'] = pd.to_datetime(df['date'])

# 添加半年度欄位
df['half_year'] = df['date'].dt.month.apply(lambda x: 'H1' if x <= 6 else 'H2')

# 添加年份欄位
df['year'] = df['date'].dt.year

# 按 app、year 和 half_year 分組，計算每個分組的評論數量
review_counts = df.groupby(['app', 'year', 'half_year']).size().unstack(fill_value=0).reset_index()

# 將數據轉換為折線圖格式
review_counts['period'] = review_counts['year'].astype(str) + " " + review_counts['half_year']
review_counts = review_counts.pivot(index='period', columns='app', values=0).fillna(0)

# 按時間順序排序 period
review_counts = review_counts.sort_index()

# 繪製折線圖，把五個 app 的數據分佈都呈現在一張圖裡
plt.figure(figsize=(14, 8))

# 每個 app 一條折線
for app in review_counts.columns:
    plt.plot(review_counts.index, review_counts[app], marker='o', linestyle='-', label=app)

# 添加標題和標籤
plt.title('Review Counts for Different Apps (H1 vs H2)')
plt.xlabel('Year and Half Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')

# 添加圖例
plt.legend(title='App', loc='upper left')

# 顯示圖表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 檔案
df = pd.read_csv('new_result/all_concat.csv')

# 確保 'date' 欄位是 datetime 類型
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 添加半年度欄位
df['half_year'] = df['date'].dt.month.apply(lambda x: 'H1' if x <= 6 else 'H2')

# 添加年份欄位
df['year'] = df['date'].dt.year

# 按 app、year 和 half_year 分組，計算每個分組的評論數量
review_counts = df.groupby(['app', 'year', 'half_year']).size().reset_index(name='counts')

# 將數據轉換為折線圖格式
review_counts['period'] = review_counts['year'].astype(str) + " " + review_counts['half_year']
pivot_table = review_counts.pivot(index='period', columns='app', values='counts').fillna(0)

# 按時間順序排序 period
pivot_table = pivot_table.sort_index()

# 計算累積總和
cumulative_pivot = pivot_table.cumsum()

# 繪製堆積折線圖（累積的方式）
plt.figure(figsize=(14, 8))

# 繪製每個 app 的累積數據
for app in cumulative_pivot.columns:
    plt.plot(cumulative_pivot.index, cumulative_pivot[app], marker='o', linestyle='-', label=app)

# 添加標題和標籤
plt.title('Cumulative Review Counts for Different Apps (H1 vs H2)')
plt.xlabel('Year and Half Year')
plt.ylabel('Cumulative Number of Reviews')
plt.xticks(rotation=45, ha='right')

# 添加圖例
plt.legend(title='App', loc='upper left')

# 顯示圖表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('new_result/all_concat.csv')

# 确保 'date' 列是 datetime 类型
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 添加半年度列
df['half_year'] = df['date'].dt.month.apply(lambda x: 'H1' if x <= 6 else 'H2')

# 添加年份列
df['year'] = df['date'].dt.year

# 按 app、year 和 half_year 分组，计算每个分组的评论数量
review_counts = df.groupby(['app', 'year', 'half_year']).size().reset_index(name='counts')

# 创建合并年份和半年的 period 列
review_counts['period'] = review_counts['year'].astype(str) + " " + review_counts['half_year']

# 按时间顺序排列 period
pivot_table = review_counts.pivot(index='period', columns='app', values='counts').fillna(0)
pivot_table = pivot_table.sort_index()

# 获取年份列表用于显示在 x 轴上
years = review_counts['year'].unique()

# 绘制折线图
plt.figure(figsize=(14, 8))

# 每个 app 一条折线，使用默认颜色
for app in pivot_table.columns:
    plt.plot(pivot_table.index, pivot_table[app], marker='o', linestyle='-', label=app)

# 修改 x 轴标签为年份，并保持水平显示
plt.xticks(ticks=range(0, len(pivot_table.index), 2), labels=years)

# 添加标题和标签
plt.title('Review Counts for Different Apps')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')

# 添加图例
plt.legend(title='App', loc='upper left')

# 显示图表
plt.show()
```

```python
# 生成箱线图
plt.rcParams['figure.figsize'] = [6, 6.5]
boxplot = df.boxplot(column='subjectivity', by='app', grid=False)

plt.title('User Subjectivity for Different Swimming Training Apps')
plt.xlabel('App')
plt.ylabel('Subjectivity')
plt.suptitle('')
plt.xticks(rotation=0)  # 确保应用程序名称是水平的

# 获取每个箱线图的五数总结数据（最小值、第一四分位数、中位数、第三四分位数、最大值）
grouped = df.groupby('app')['subjectivity']
stats = grouped.describe()

adjustments = {
    'duoswim': 0.04,
    'form-swim': 0.06,
    'myswimpro': 0.06,
    'swim-com': 0.06,
    'swimup': 0.06
}

for i, app in enumerate(stats.index, 1):  # enumerate 从1开始，因为箱线图的箱子从1开始
    mean = stats.loc[app, 'mean']
    adjustment = adjustments[app]
    
    # 调整标签的位置和字体
    plt.text(i, mean + adjustment, f'{mean:.2f}', horizontalalignment='center', verticalalignment='top', fontsize=10, color='green', bbox=dict(facecolor='none', edgecolor='none', alpha=0.5))

# 添加显著性标记
def add_significance(ax, x1, x2, y, text):
    x = [x1, x1, x2, x2]
    y_list = [y, y + 0.01, y + 0.01, y]
    ax.plot(x, y_list, lw=1.5, color='black')
    ax.text((x1 + x2) * .5, y + 0.015, text, ha='center', va='bottom', color='black', fontsize=10)

ax = plt.gca()
y_max = df['subjectivity'].max() + 0.1
add_significance(ax, 2, 3, y_max, '*')
add_significance(ax, 2, 4, y_max + 0.05, '**')

plt.show()
```
