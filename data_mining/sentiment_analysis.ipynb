```python
import pandas as pd
from textblob import TextBlob
```

```python
def analyze_sentiment(df):
    df['polarity'] = df['content'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
    df['subjectivity'] = df['content'].apply(lambda x: TextBlob(str(x)).sentiment.subjectivity)
    df['sentiment'] = df['polarity'].apply(lambda x: 'positive' if x > 0 else ('neutral' if x == 0 else 'negative'))
    return df

def analysis_to_csv(app: str):
    reviews = pd.read_csv(f'result/{app}/{app}_combined.csv')
    reviews = analyze_sentiment(reviews)
    if 'Unnamed: 0' in reviews.columns:
        reviews = reviews.drop(columns=['Unnamed: 0'])
    reviews.to_csv(f'{app}_sentiment.csv', index=False)
    return reviews
```

```python
df = analysis_to_csv('duoswim')
df
```
