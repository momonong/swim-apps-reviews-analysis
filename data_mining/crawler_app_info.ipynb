```python
import pandas as pd
from google_play_scraper import app
```

## Google

```python

id_list = ['com.duo.duoswim',
           'com.formathletica.formandroid',
           'com.myswimpro.android.app',
           'com.spiraledge.swimapp',
           'io.swimup.application']

for id in id_list:
    app_info = app(
        id,
        lang='en',  
        country='us'  
    )
    print(app_info['title'])
```

```python
app_info = app(
    id_list[2],
    lang='en',  
    country='us'  
)
app_info
```

```python
info_list = {}
for id in id_list:
    app_info = app(
        id,
        lang='en',  
        country='us'  
    )
    info_list[app_info['title']] = app_info
```

```python
info_df = pd.DataFrame(info_list)
info_df = info_df.T
info_df.drop(columns="title", axis=1, inplace=True)
info_df.to_csv('app_info_google.csv')
info_df
```
