```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data.csv')
df = df.drop(columns=['Relay?'])
df = df.drop(columns=['Team'])
df = df.drop(columns=['Gender'])
df = df.drop(columns=['Year'])
df['Results'] = df['Results'] - 3.8
df['Results'] = df['Results'].round(3)
df['Athlete'] = '林姵彣'
df.rename(columns={'Distance (in meters)': 'Distance (m)'}, inplace=True)
df
```

```python
import pandas as pd
import numpy as np

# 设置随机种子以确保结果的可复现性
np.random.seed(42)

# 创建日期范围
dates = pd.date_range(start="2021-01-01", periods=10, freq='M')

# 生成数据
data = {
    "Height (cm)": [174] * 10, 
    "Weight (kg)": [61.0] * 10,    
    "Trainging hours": np.random.normal(3, 0.25, 10).round(2),  # 训练时间假设平均4小时，标准差1小时
    "Calories (kcal)": np.random.normal(2500, 500, 10).round(2),  # 日均卡路里摄入2500千卡，标准差500千卡
    "Dates": dates,
}

# 转换为DataFrame
athlete_data = pd.DataFrame(data)

athlete_data['Weight (kg)'] += np.random.uniform(-1, 1, size=len(df))
athlete_data['Weight (kg)'] = athlete_data['Weight (kg)'].round(2)

# 显示生成的数据
athlete_data
```

```python
# 横向拼接
data = pd.concat([athlete_data, df], axis=1)
```

```python
# 指定新的列顺序
new_order = [
    "Athlete",
    "Height (cm)", "Weight (kg)",
    "Trainging hours", "Calories (kcal)",
    "Event", "Dates", "Location", "Distance (m)", "Stroke", "Results"
]

# 重新索引DataFrame以改变列的顺序
data = data[new_order]

# 为了可复现性设置随机种子
np.random.seed(6)

# 随机分配新的地点
new_locations = np.random.choice(['Tainan', 'Taichung', 'Taipei'], size=len(data), replace=True)
data['Location'] = new_locations

data.to_csv('sample_data2.csv', index=False)
data
```

```python
# TODO
# training hours, calories 改成 distance and tensity, heart rate
# add sleeping tests, which need to written in 0-56 scale (0-28, 29-42, 43-56)
# 

import pandas as pd
import numpy as np

df = pd.read_csv('sample_data2.csv')

# 更改列名
df.rename(columns={
    'Trainging hours': '訓練距離',
    'Calories (kcal)': '強度'
}, inplace=True)

# 加入心率列，随机生成心率数据
np.random.seed(42)  # 确保可复现性
df['心率'] = np.random.randint(60, 100, size=len(df))  # 假设心率在60到100之间

# 添加睡眠问卷量表分数列，随机生成0-56的数值
df['睡眠問卷分數'] = np.random.randint(0, 57, size=len(df))

# 根据睡眠问卷分数添加等级
conditions = [
    (df['睡眠問卷分數'] <= 28),
    (df['睡眠問卷分數'] <= 42),
    (df['睡眠問卷分數'] <= 56)
]
choices = [1, 2, 3]
df['睡眠等級'] = np.select(conditions, choices)

df.to_csv('sample_data3.csv', index=False)
df
```

```python
df.columns = ['運動員', '身高(cm)', '體重(kg)', '訓練距離', '賽事', '日期', 
              '地點', '距離(m)', '泳姿', '成績', '心率', '睡眠問卷分數', '睡眠等級']

# 每周训练天数为3天，每天跑2公里
hours_to_km_per_week_factor = 3 * 2  # 一周总共跑的公里数

# 转换训练距离：我们假设原始数据中的“訓練距離”列现在表示的是每周训练的天数，现在将其转换为公里数
df['訓練距離'] = df['訓練距離'] * hours_to_km_per_week_factor

# 更改列名以反映新单位
df.rename(columns={'訓練距離': '訓練距離(km/週)'}, inplace=True)
# df.drop(columns=['卡路里'], inplace=True)

# 设置随机种子以确保结果的可复现性
np.random.seed(42)

# 生成更真实的训练心率，范围在120到180次/分钟
df['心率'] = np.random.randint(120, 181, size=len(df))

df.to_csv('sample_data3.csv', index=False)
df
```

```python
import pandas as pd

df = pd.read_csv('sample_data3.csv')
# 替换地点名称为赛事级别
location_to_competition_level = {
    'Taipei': '全國級',
    'Taichung': '縣市級',
    'Tainan': '國際級'
}

df['地點'] = df['地點'].replace(location_to_competition_level)
df.rename(columns={'地點': '賽事等級'}, inplace=True)
df['訓練距離(km/週)']  = (df['訓練距離(km/週)']/100000).round(2)
df.to_csv('sample_data4.csv', index=False)

df
```
