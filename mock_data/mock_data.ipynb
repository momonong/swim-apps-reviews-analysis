```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data.csv')
df = df.drop(columns=['Relay?'])
df = df.drop(columns=['Team'])
df = df.drop(columns=['Gender'])
df = df.drop(columns=['Year'])
df['Results'] = df['Results'] - 3.8
df['Results'] = df['Results'].round(3)
df['Athlete'] = '林姵彣'
df.rename(columns={'Distance (in meters)': 'Distance (m)'}, inplace=True)
df
```

```python
import pandas as pd
import numpy as np

# 设置随机种子以确保结果的可复现性
np.random.seed(42)

# 创建日期范围
dates = pd.date_range(start="2021-01-01", periods=10, freq='M')

# 生成数据
data = {
    "Height (cm)": [174] * 10, 
    "Weight (kg)": [61.0] * 10,    
    "Trainging hours": np.random.normal(3, 0.25, 10).round(2),  # 训练时间假设平均4小时，标准差1小时
    "Calories (kcal)": np.random.normal(2500, 500, 10).round(2),  # 日均卡路里摄入2500千卡，标准差500千卡
    "Dates": dates,
}

# 转换为DataFrame
athlete_data = pd.DataFrame(data)

athlete_data['Weight (kg)'] += np.random.uniform(-1, 1, size=len(df))
athlete_data['Weight (kg)'] = athlete_data['Weight (kg)'].round(2)

# 显示生成的数据
athlete_data
```

```python
# 横向拼接
data = pd.concat([athlete_data, df], axis=1)
```

```python
# 指定新的列顺序
new_order = [
    "Athlete",
    "Height (cm)", "Weight (kg)",
    "Trainging hours", "Calories (kcal)",
    "Event", "Dates", "Location", "Distance (m)", "Stroke", "Results"
]

# 重新索引DataFrame以改变列的顺序
data = data[new_order]

# 为了可复现性设置随机种子
np.random.seed(6)

# 随机分配新的地点
new_locations = np.random.choice(['Tainan', 'Taichung', 'Taipei'], size=len(data), replace=True)
data['Location'] = new_locations

data.to_csv('sample_data2.csv', index=False)
data
```
