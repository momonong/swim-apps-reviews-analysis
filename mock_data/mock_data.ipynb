```python
import numpy as np

df1 = pd.read_csv('sample_data4.csv')
df2 = pd.read_csv('world.csv')
# 合并两个DataFrame
df1 = df1.rename(columns={"泳姿": "項目", "距離": "水道距離"})
df2 = df2.rename(columns={"Event": "項目", "Time": "成績", "Competition": "賽事", "Date": "日期"})


# 假设缺失的数据主要涉及心率和睡眠问卷分数，这里随机生成这些数据
heart_rates = np.random.randint(120, 180, size=8)  # 随机生成心率，范围120-180
sleep_scores = np.random.randint(1, 56, size=8)  # 随机生成睡眠问卷分数，范围1-55
average_distance = 8.5  # average training distance per week in km
new_distances = average_distance + np.random.uniform(-0.5, 0.5, size=8)
df2['心率'] = heart_rates
df2['睡眠問卷分數'] = sleep_scores
df2['訓練距離(km/週)'] = new_distances.round(2)
df2['賽事等級'] = "國際級"
# 根据睡眠问卷分数添加等级
conditions = [
    (df2['睡眠問卷分數'] <= 28),
    (df2['睡眠問卷分數'] <= 42),
    (df2['睡眠問卷分數'] <= 56)
]
choices = [1, 2, 3]
df2['睡眠等級'] = np.select(conditions, choices)
df1['日期'] = pd.to_datetime(df1['日期'], format='%Y-%m-%d')
df2['日期'] = pd.to_datetime(df2['日期'], format='%Y-%m-%d')
df1['日期'] = df1['日期'].apply(lambda x: x.replace(year=2023))

df = pd.concat([df1, df2], ignore_index=True)
df.drop(columns=['運動員'], inplace=True)
df.drop(columns=['身高(cm)'], inplace=True)
df.drop(columns=['距離(m)'], inplace=True)
df.drop(columns=['體重(kg)'], inplace=True)

# Calculate event distribution from the lower half (indexes 10 to 19)
event_distribution = df.iloc[10:]['項目'].value_counts(normalize=True)

# Apply the distribution to the upper half
for i in range(10):
    selected_event = np.random.choice(event_distribution.index, p=event_distribution.values)
    df.at[i, '項目'] = selected_event
    # Ensure the score matches the selected event
    selected_scores = df[df['項目'] == selected_event]['成績'].values
    df.at[i, '成績'] = np.random.choice(selected_scores)
    
df.to_csv('sample_data5.csv')
df2
```
