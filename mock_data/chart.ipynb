```python
import matplotlib.pyplot as plt

# Data setup
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
heights = [160, 161, 161, 162]  # in cm
heart_rates = [75, 73, 72, 70]  # in bpm
swim_times = [35, 34, 33, 32]  # in seconds

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# First axis
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Height (cm)', color='tab:blue')
ax1.plot(quarters, heights, label='Height (cm)', color='tab:blue', marker='o')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Second axis (share the same x-axis)
ax2 = ax1.twinx()
ax2.set_ylabel('Heart Rate (bpm)', color='tab:red')
ax2.plot(quarters, heart_rates, label='Heart Rate (bpm)', color='tab:red', marker='o')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Third axis (new figure for clarity)
fig2, ax3 = plt.subplots(figsize=(10, 6))
ax3.set_xlabel('Quarter')
ax3.set_ylabel('50m Freestyle Swim Time (seconds)', color='tab:green')
ax3.plot(quarters, swim_times, label='50m Freestyle Time (s)', color='tab:green', marker='o')
ax3.tick_params(axis='y', labelcolor='tab:green')

# Titles and legend
ax1.set_title('Height and Heart Rate over Quarters')
ax3.set_title('50m Freestyle Swim Time over Quarters')

# Show plots
plt.show()
```
